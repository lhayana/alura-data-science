```python
# Selecione somente os imóveis classificados com tipo 'Apartamento'.

selecao_ap = dados['Tipo'] == 'Apartamento'

aps = dados[selecao_ap]
aps
```

```python
# Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.
casas_tipos = ['Casa', 'Casa de Condomínio', 'Casa de Vila'];
selecao_casas =  dados['Tipo'].isin(casas_tipos)
casas = dados[selecao_casas]
casas
```

```python
# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

limites = ((dados['Area'] > 60) & (dados['Area'] > 100))
imoveis_limites = dados[limites]
imoveis_limites
```

```python
# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.

limites2 = ((dados['Valor'] < 2000) & (dados['Quartos'] >= 4))
imoveis_limites2 = dados[limites2]
imoveis_limites2
```

```python
print(aps.shape[0], casas.shape[0], imoveis_limites.shape[0], imoveis_limites2.shape[0])
```
